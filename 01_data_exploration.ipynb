```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sentimentdataset (1).csv")
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"], errors="ignore")
```

```python
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].str.strip()
```

```python
plt.figure(figsize=(12, 60))  # Set figure size

sns.countplot(data=df, y="Sentiment", hue="Platform", palette="pastel")

plt.title("Sentiment Distribution by Platform", fontsize=16)
plt.xlabel("Count", fontsize=14)
plt.ylabel("Sentiment", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x="Platform", order=df['Platform'].value_counts().index)
plt.title("Posts per Platform")
plt.xticks(rotation=45)
plt.show()
```
